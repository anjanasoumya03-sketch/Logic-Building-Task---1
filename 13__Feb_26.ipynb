```python
# 2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nPass / Fail Analysis:")
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)

```

```python
# 3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    # Remove extra spaces and convert to lowercase
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("\nCleaned Names List:")
print(cleaned_names)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis:")

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    # Flag messages longer than 10 characters
    if length > 10:
        print("-> This message is longer than 10 characters.")

```

```python
# 5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nError Log Analysis:")
print("Total ERROR entries:", error_count)
```
